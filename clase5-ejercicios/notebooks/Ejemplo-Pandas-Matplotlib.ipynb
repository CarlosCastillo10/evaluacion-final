```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy as sp

```

```python
x = range(100)
y = x
plt.grid(True, linewidth=2)
plt.xlabel("valores de las (x)", fontsize = 20)
plt.ylabel("valores de las (y)", fontsize = 20)
plt.title(u"Título de la gráfica", fontsize = 30)
plt.text(20,20, u"Más texto", fontsize=14)
a = plt.plot(x,y, 'g--')
```

```python
#Ejemplo
x = sp.linspace(0, 5, 100)
y = sp.sin(x)
plt.figure()
plt.title(u"Título de la gráfica", fontsize = 25)
plt.xlabel("valores de las (x)", fontsize = 20)
plt.ylabel("valores de las (y)", fontsize = 20)
plt.plot(x,y)
plt.show()

```
